```python
import pandas as pd
%matplotlib inline 
#Needed to inline graphs
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('Chennai_PP_data.csv')
```

```python
degrees = data.iloc[:,1::2]
degrees
```

```python
#plt.plot(data.iloc[:,0])
plt.plot(degrees.iloc[:,1],label='jupiter')
#plt.plot(degrees.iloc[:,3],label='mars')
#plt.plot(degrees.iloc[:,4],label='mercury')
#plt.plot(degrees.iloc[:,5],label='moon')
plt.plot(degrees.iloc[:,6],label='neptune')
plt.plot(degrees.iloc[:,8],label='Saturn')
plt.plot(degrees.iloc[:,-1],label='Venus')
plt.plot(degrees.iloc[:,-2],label='Ur')




plt.legend()
plt.show()
```
