## Encoding Categorical label

### Recap

```python


#Supress warnings
import warnings
warnings.filterwarnings('ignore')

# Importing the libraries
import numpy as np
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 3].values

# Taking care of missing data
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

### Encoding to integers using Label Encoder

```python
#Encode the country and purchased variables
from sklearn.preprocessing import LabelEncoder
help(LabelEncoder)
```

```python
le_X = LabelEncoder() #create obj 
```

```python
le_X.fit_transform(X[:,0]) #you can also use fit_transform menthod for imputer
```

```python
X[:,0] = le_X.fit_transform(X[:,0])
X
```

#### However, this is not desirable as ML model thinks one country could be better than other. So we need 01 encodings

```python
#for 01 encoding
from sklearn.preprocessing import OneHotEncoder
help(OneHotEncoder)
```

```python
oe = OneHotEncoder(categorical_features=[0])#specify index of column that has categorical variable
```

```python
X = oe.fit_transform(X).toarray()
```

```python
np.set_printoptions(precision=4)
np.set_printoptions(suppress=True) #Otherwise prints in scientific format
print(X)
```
