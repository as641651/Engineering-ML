## Decision Tree Classifier


### 1) Get the dataset

```python
#Supress warnings
import warnings
warnings.filterwarnings('ignore')

# Importing the libraries
import numpy as np
import pandas as pd

np.set_printoptions(precision=3)
np.set_printoptions(suppress=True) #Otherwise prints in scientific format

%matplotlib inline 
#Needed to inline graphs
import matplotlib.pyplot as plt


# Importing the dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
print("No. of datapoints : ", len(dataset))
dataset[:10]
```

### 2) Separate the dep and indep variables into a numpy array

We are only going to consider **age** and **salary** as independent variables to predict purchases.

```python
X = dataset.iloc[:,[2,3]].values 
y = dataset.iloc[:, 4].values
```

### 3) Splitting into training and test set

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

### 4) Feature scaling

Is feature scaling needed for this algorithm? **NO**. Decision trees are **not based on eucleadian diatance**. But they are useful to generate high resolution plots

```python
# # Feature Scaling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_trainS = sc.fit_transform(X_train)
# X_testS = sc.transform(X_test)
```

### 5) Predict using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier as DC
help(DC)
```

```python
classifier = DC(criterion='entropy')
classifier
```

```python
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
pd.DataFrame(np.hstack((X_test, y_pred.reshape(-1,1),y_test.reshape(-1,1))),columns=["Age","Salary","Purch_predict","Purch_actual"])[:10] #hstack takes 2D arrays
```

### 6) Printing Confusion Matrix

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

### 7) Visualizing the class boundry

(Without features scaled)

```python
import plot_binary as pb

plt = pb.plot_binary(X_train,y_train,classifier,1)
plt.title('Decision Tree Classifier (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()

plt = pb.plot_binary(X_test,y_test,classifier,1)
plt.title('Decision Tree Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

**Notice :** This training captures even those points are sandwiched between different classes. This indicates **overfitting**
